```python
#updating columns name
df.columns=['state','date','frequency','estimated unemployment rate','estimated employed','estimated labour participation rate','region','longitude','latitude']
```

```python
df['date']=pd.to_datetime(df['date'],dayfirst=True)
```

```python
df.info()
```

```python
df.head(20)
```

```python
#we can also extract month from date for better analysis
df['month']=df['date'].dt.month_name()
df.head()
```

Visualization

```python
#Number of employees according to different regions of India:
df.columns= ["state","date","frequency",
               "estimated unemployment rate","estimated employed",
               "estimated labour participation rate","region",
               "longitude","latitude","month"]
plt.title("Indian Unemployment in 2020")
sns.histplot(x="estimated employed", hue="region", data=df)
plt.show()
```

Region wise Unemployement details

```python
region_stats = df.groupby(['region'])[['estimated unemployment rate', 'estimated employed', 
                                       'estimated labour participation rate']].mean().reset_index()
round(region_stats, 2)
```

```python
#Unemployment rate according to different regions of India
plt.figure(figsize=(10, 10))
plt.title("Indian Unemployment rate")
sns.histplot(x="estimated unemployment rate", hue="region", data=df)
plt.show()
```

```python
# Unemployment Rate by Regions

plt.bar(df['region'], df["estimated unemployment rate"])
plt.xlabel('Regions in India')
plt.ylabel('Estimated unemployment rate')
plt.show()
```
